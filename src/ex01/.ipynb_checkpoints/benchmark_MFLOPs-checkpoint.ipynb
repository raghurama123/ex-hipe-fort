```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
from sklearn.metrics import r2_score
```

```python
mflops=pd.read_csv('benchmark_MFLOPS.csv',header=0)
mflops.head()
```

```python
plt.rcParams["figure.figsize"] = (8,6)

fig, ax = plt.subplots()

ax.set_xlabel("Vector length")
ax.set_ylabel("MFLOP/sec")

x=mflops['N'].values

N=8
for i in range(1,N):
    y=mflops[mflops.keys()[i]].values
    ax.loglog(x,y,'o-')

ax.legend(mflops.keys()[1:N])

fig.savefig('benchmarks_MLOPS.pdf')
```
